```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf
```

```python
# Import our input dataset
df = pd.read_csv("charity_data.csv")
df.head()
```

```python
#remove name and EIN:
charity_clean_df=df.drop(columns=["EIN","NAME"])
charity_clean_df
```

```python
# Generate our categorical variable list:
charity_cat = charity_clean_df.dtypes[charity_clean_df.dtypes == "object"].index.tolist()
```

```python
# Check the number of unique values in each column:
charity_clean_df[charity_cat].nunique()
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(charity_clean_df[charity_cat]))

# Add the encoded variable names to the DataFrame
encode_df.columns = enc.get_feature_names(charity_cat)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
charity_df = charity_clean_df.merge(encode_df,left_index=True, right_index=True)
charity_df = charity_df.drop(charity_cat,1)
charity_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = charity_df["IS_SUCCESSFUL"].values
X = charity_df.drop(["IS_SUCCESSFUL"],1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```
